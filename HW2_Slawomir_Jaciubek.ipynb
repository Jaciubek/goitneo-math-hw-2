###Task 1. Solving matrices
Solve the system of linear algebraic equations in matrix form using the inverse matrix:

###a)

\begin{cases}
-x_1 + x_2 + 2x_3 = 1 \\
-x_2 - 3x_3 = -4 \\
4x_1 - 3x_2 + 2x_3 = 7
\end{cases}

```python
import numpy as np

# matrix a
matrix_a = np.array([[-1, 1, 2],
              [0, -1, -3],
              [4, -3, 2]])

# matrix b
matrix_b = np.array([1, -4, 7])

# inverse of matrix a
matrix_a_inv = np.linalg.inv(matrix_a)

# multiplying a_inv by matrix b
x = np.dot(matrix_a_inv, matrix_b)

# Round the values
x_rounded = np.round(x).astype(int)

# Printing results
print(f"The solutions to the system of equations are the following numbers: x1 = {x_rounded[0]}, x2 = {x_rounded[1]}, x3 = {x_rounded[2]}")
```

###b)

$$
\begin{cases}
3x_1 - x_2 = 4 \\
2x_1 - 2x_2 + 4x_3 = 8 \\
5x_1 - 6x_2 + 2x_3 = 13
\end{cases}
$$

```python
import numpy as np

matrix_c = np.array([[3, -1, 0],
              [2, -2, 4],
              [5, -6, 2]])

matrix_d = np.array([4, 8, 13])

matrix_c_inv = np.linalg.inv(matrix_c)

x = np.dot(matrix_c_inv, matrix_d)

x_rounded = np.round(x).astype(int)

print(f"The solutions to the system of equations are the following numbers: x1 = {x_rounded[0]}, x2 = {x_rounded[1]}, x3 = {x_rounded[2]}")
```

###Task 3. Solving a problem with a textual condition
Three types of specialists worked on the project: designers, developers, and testers. We know the number of hours spent at each stage of development and the amount of money allocated for each stage as cost price.

The entire development was divided into 3 stages by the manager: creating the design, developing the software, and final testing.

The first stage took the designer 20 hours and the developers 3 hours; this stage was allocated a budget of 800 dollars. The second stage took the designers 8 hours, developers 28 hours, and testers 4 hours and was allocated 900 dollars. In the final stage, designers worked for 4 hours, developers for 12 hours, and testers for 32 hours. The budget was 600 dollars.

Find the cost per hour for each specialist by solving a system of linear equations, based on the data from the development stages.

###SOLUTION

To find the cost per hour for each specialist, we solve the system of linear equations:

For the first stage:
$$ 20 \cdot \text{{designer rate}} + 3 \cdot \text{{developer rate}} = 800 $$

For the second stage:
$$ 8 \cdot \text{{designer rate}} + 28 \cdot \text{{developer rate}} + 4 \cdot \text{{tester rate}} = 900 $$

For the final stage:
$$ 4 \cdot \text{{designer rate}} + 12 \cdot \text{{developer rate}} + 32 \cdot \text{{tester rate}} = 600 $$

The solution to these equations gives us the hourly rates for each specialist:
- Designer rate: $( \frac{1070}{29} )$ dollars per hour ~ $36,8965...$,
- Developer rate: $( \frac{600}{29} )$ dollars per hour ~ $20,6896...$,
- Tester rate: $( \frac{185}{29} )$ dollars per hour ~ $6,3793...$.

We can create the following system of equations:

$$
\begin{cases}
20x + 3y = 800, \\
8x + 28y + 4z = 900, \\
4x + 12y + 32z = 600
\end{cases}
$$

```python
import numpy as np

# matrix g
matrix_g = np.array([[20, 3, 0],
              [8, 28, 4],
              [4, 12, 32]])
# matrix h
matrix_h = np.array([800, 900, 600])

result_matrix = np.linalg.solve(matrix_g, matrix_h)
print(f"The solution to the above problem is as follows: Cost/hour of designers work: {result_matrix[0]}, Cost/hour of developers work: {result_matrix[1]} and Cost/hour of testers work: {result_matrix[2]}")
```

###Task
The amount of 100 z was spent in coins of 1, 2, and 5 zoty - a total of 32 coins, with the number of 5 zoty coins being 2 more than twice the number of 1 zoty coins. How many coins of each type were spent?:

Let's define the variables: \\
$( x )$ - the number of 1 zoty coins, \\
$( y )$ - the number of 2 zoty coins, \\
$( z )$ - the number of 5 zoty coins.

A total of 32 coins of each type were spent. This can be expressed using the equation: \\
 +  +  = 32

Taking into account the denominations of the coins, the equation can be written as: \\
 + 2 + 5 = 100

The information that the number of 5 zoty coins was 2 more than twice the number of 1 zoty coins can be expressed in the equation: \\
 = 2 + 2 -------> 2 -  = -2 \\

The equations describing the situation are:
$$
\begin{cases}
  +  +  = 32 \\
  + 2 + 5 = 100 \\
 2 -  = -2
\end{cases}
$$

```python
import numpy as np

matrix_i = np.array(
 [[1, 1, 1],
 [1, 2, 5],
 [2, 0, -1]]
 )

matrix_j = np.array([32, 100, -2])

result = np.linalg.solve(matrix_i, matrix_j)
rounded_result = np.round(result).astype(int)

print(rounded_result)
print(f"The solution to the system of equations is as follows: the number of 1 zoty coins = {rounded_result[0]}, the number of 2 zoty coins = {rounded_result[1]}, and the number of 5 zoty coins = {rounded_result[2]}")
```
